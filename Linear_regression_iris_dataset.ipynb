```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

```

```
iris = load_iris()

x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)

print(x)
print(20*"-----")
print(y)
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
k= 20*"-----"

print("X-Training Set \n {} \n {}".format(x_train, k))
print("X-Test Set \n {} \n {}".format(x_test, k))
print("Y-Training Set \n {} \n {}".format(y_train, k))
print("Y-Test Set \n {} \n {}".format(y_test, k))
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

k= 20*"-----"
print("X-Training Set \n {} \n {}".format(x_train, k))
print("X-Test Set \n {} \n {}".format(x_test, k))
```

```
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```
print(model.predict(scaler.transform(pd.DataFrame([[3,2,10,0]], columns=iris.feature_names))))
```

```
y_pred = model.predict(x_test)

print(y_pred)
print(y_test.values)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

```
report = classification_report(y_test, y_pred, target_names=iris.target_names)
print(report)
```
